```python
import sys
sys.path.append('/home/samuel/Dropbox (uwamath)/muscle/MuscleModel/multifil/')
from multifil.aws.run import *
%pylab inline
pylab.rcParams['figure.figsize'] = (12, 8)
import numpy as np
sys.path.append('/home/samuel/Dropbox (uwamath)/muscle/MuscleModel/Python_Codes')
sys.path.append('/Users/samuelrudy/Dropbox (uwamath)/muscle/MuscleModel/Python_Codes')
from SINDy3 import SINDy
from scipy import sparse
import itertools
import operator
import pandas as pd
from zipfile import ZipFile
import ujson
from scipy.integrate import odeint
from DMD import *

def MovingAve(x, m):
    n = len(x)
    xs = np.zeros(n-2*m)
    for j in range(n-2*m):
        xs[j] = np.mean(x[j:j+2*m])
    return xs

def ProjZeroOne(x):
    condlist = [x<0, (x>0) & (x<=1), x>1]
    choicelist = [0,x, 1]
    return np.select(condlist, choicelist)
```

## Load data

```python
sarc_list = {}

for dirpath, dirnames, filenames in os.walk('../Data/October31/'):
    for file in filenames:
        if 'data.json' in file:
            
            # get meta info
            path = os.path.join('../Data/October31/', file.split('.')[0]+'.meta.json')
            meta = pd.read_json(path)
            description = [float(j) for j in meta['actin_permissiveness_args'][0][1:-1].split(',')]    
            period = description[0]
            phase = description[1]
            half_life = description[4]
            
            # read in data
            path = os.path.join('../Data/October31/', file)
            if (period, phase, half_life) in sarc_list:
                sarc_list[(period, phase, half_life)].append(pd.read_json(path))
            else: sarc_list[(period, phase, half_life)] = [pd.read_json(path)]

print(sarc_list.keys())
print(sarc_list[(50.0, 0.5, 2.5)][0].keys())
```

## Linear function for axial force and radial tension

$$
\begin{pmatrix}
\text{axial force}\\
\text{radial tension}
\end{pmatrix} = B \begin{pmatrix}
\text{fraction free}\\
\text{fraction loose}\\
\text{fraction tight}
\end{pmatrix} + C\begin{pmatrix}
z \\
ls\\
perm
\end{pmatrix}
$$

```python
dt = 0.1
data =[]

for key, runs in sarc_list.items():
    
    period, phase, half_life = key
        
    # What are we considering to be dynamic variables and what is forcing?
    data_keys = ['axial_force', 'radial_tension']
    fraction_keys = ['xb_fraction_free','xb_fraction_loose', 'xb_fraction_tight']
    forcing_keys = ['z_line', 'lattice_spacing', 'actin_permissiveness']

    # Toss the first period
    start = int(period/(2*dt))
    end = int(2000.0*dt/period - 0.5)*int(period/dt) + start
    n = end-start-1

    X = np.zeros((len(data_keys),end-start-1))
    S = np.zeros((len(fraction_keys),end-start-1))
    G = np.zeros((len(forcing_keys),end-start-1))

    for run in runs:

        # center lattice spacing and z line
        run['z_line'] = run['z_line'] - np.mean(run['z_line'])
        run['lattice_spacing'] = run['lattice_spacing'] - np.mean(run['lattice_spacing'])

        X = X + np.vstack([run[j][start:end-1].reshape(1,n) for j in data_keys])
        S = S + np.vstack([run[j][start:end-1].reshape(1,n) for j in fraction_keys])
        g = np.vstack([run[j][start:end-1].reshape(1,n) for j in forcing_keys])
#         g = np.vstack([g, np.array([run['z_line'][j+1] - run['z_line'][j] for j in np.arange(start,end-1)]).reshape(1,n)])
        G = G + g

    X = X/len(runs)
    S = S/len(runs)
    G = G/len(runs)

    data.append([X,S,G])

X1 = np.hstack(dataset[0][:,1:] for dataset in data)
S = np.hstack(dataset[1][:,1:] for dataset in data)
G = np.hstack(dataset[2][:,1:] for dataset in data)

RHS = np.vstack([S,G])
U,S,Vt = np.linalg.svd(RHS, full_matrices = False)

BC = X1.dot(Vt.T).dot(diag(S**-1)).dot(U.T)
B = BC[:,:3]
C = BC[:,3:]
```

## Learn transition probabilities as quadratic funcitons of forcing

```python
N = 720

P12 = []
P11 = []
P22 = []
P21 = []
P23 = []
P33 = []
P31 = []
P32 = []

for key, runs in sarc_list.items():
    
    for run in runs:
        
        n = len(run)
        zm = 1250#np.mean(run['z_line'])
        
        N1 = rint(run['xb_fraction_free']*N)
        N2 = rint(run['xb_fraction_loose']*N)
        N3 = rint(run['xb_fraction_tight']*N)
        
        # Loop over timesteps
        for j in range(1,n):
            
            perm = run['actin_permissiveness'][j]
            zt = np.exp(-abs(run['z_line'][j] - run['z_line'][j-1]))
            z = run['z_line'][j]-zm
            
            # Only look at ones where we can get transition probabilities
            if N1[j] != 0:
                P12.append([perm, zt, z, run['xb_trans_12'][j]/N1[j]])
                P11.append([perm, zt, z, 1 - run['xb_trans_12'][j]/N1[j]])
            
            if N2[j] != 0:
                P21.append([perm, zt, z, run['xb_trans_21'][j]/N2[j]])
                P23.append([perm, zt, z, run['xb_trans_23'][j]/N2[j]])
                P22.append([perm, zt, z, 1-run['xb_trans_21'][j]/N2[j]-run['xb_trans_23'][j]/N2[j]])
            
            if N3[j] != 0:
                P31.append([perm, zt, z, run['xb_trans_31'][j]/N3[j]])
                P32.append([perm, zt, z, run['xb_trans_32'][j]/N3[j]])
                P33.append([perm, zt, z, 1-run['xb_trans_31'][j]/N3[j]-run['xb_trans_32'][j]/N3[j]])
```

```python
P = {}
P[(1,2)] = P12
P[(2,1)] = P21
P[(2,3)] = P23
P[(3,2)] = P32
P[(3,1)] = P31
Xi = {}

for key, Pr in P.items():
    
    n = len(Pr)
    
    X = np.vstack(j[:3] for j in Pr)
    X = np.hstack([np.ones((n,1)),X, X**2, (X[:,0]*X[:,1]).reshape(n,1), (X[:,0]*X[:,2]).reshape(n,1), (X[:,2]*X[:,1]).reshape(n,1)])
    PP = np.vstack(j[3] for j in Pr)

    Xi[key] = np.linalg.lstsq(X,PP)[0]
```
