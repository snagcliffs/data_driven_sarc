```python
import sys

# Correct for use on various computers
sys.path.append('/home/samuel/Dropbox (uwamath)/muscle/MuscleModel/multifil/')
sys.path.append('/home/samuel/Dropbox (uwamath)/muscle/MuscleModel/Python_Codes')
sys.path.append('/Users/samuelrudy/Dropbox (uwamath)/muscle/MuscleModel/Python_Codes')

from multifil.aws.run import *
%pylab inline
pylab.rcParams['figure.figsize'] = (12, 8)
import numpy as np
from scipy import sparse
import itertools
import operator
import pandas as pd
from zipfile import ZipFile
import ujson
from scipy.integrate import odeint
import seaborn

def MovingAve(x, m):
    n = len(x)
    xs = np.zeros(n-2*m)
    for j in range(n-2*m):
        xs[j] = np.mean(x[j:j+2*m])
    return xs

def ProjZeroOne(x):
    condlist = [x<0, (x>0) & (x<=1), x>1]
    choicelist = [0,x, 1]
    return np.select(condlist, choicelist)

def PolynomialFunctions(r, P=2, max_var = 2):
    """
    Builds a default collection of functions for the rhs

    Input:
    r = dimension of inputs to polynomial functions
    P = highest order of polynomials to include
    max_var = maximum number of variables present in any monomial 

    Output:
    rhs_functions = dictionary: key=powers ---> [lambda function, string description]
    """

    powers = []
    f = lambda x, y : np.prod(np.power(list(x), list(y)))

    for p in range(P+1):
        size = r + p - 1
        for indices in itertools.combinations(range(size), r-1):
            starts = [0] + [index+1 for index in indices]
            stops = indices + (size,)
            power = tuple(map(operator.sub, stops, starts))
            if sum(1 for j in power if j != 0) <= max_var:
                powers.append(power)

    return [[lambda x, y = power: f(x,y), str('polynomial: '+str(power))] for power in powers]

def PolynomialKernel(X, P, max_var = 2):
    """
    Input: 
        X = data matrix in R^{}
        P = degree of polynomial kernel
        max_var = maximum number of terms in any monomial
    Output:
        Xp = X and each polynomial of the columns of X up to degree p
    """
    
    n,r = X.shape    
    functions = PolynomialFunctions(r,P,max_var)
    return np.hstack([np.vstack([func[0](X[j,:]) for j in range(n)]) for func in functions])
```

```python
def ElasticNet(X0, Y, lam1, lam2, hard_thresh = np.Inf, maxit = 100, normalize = 2, tol = 10**-5):
    """
    Uses accelerated proximal gradient to solve
    argmin (1/2)*||Xw-Y||_2^2 + lam_1||w||_1 + (lam_2/2)*||w||_2
    then returns w after hard threshold
    """
    
    # Obtain size of X
    n,d = X0.shape
    n2,m = Y.shape
    if n != n2: raise Exception('Dimension mismatch n != n2')
    
    # Create w
    w = np.zeros((d,m))
    w_old = np.zeros((d,m))
        
    # Initialize a few other parameters
    converge = 0
    objective = np.zeros((maxit,1))
    
    # First normalize data
    if normalize != 0:
        X = np.zeros((n,d))
        Mreg = np.zeros((d,1))
        for i in range(0,d):
            Mreg[i] = 1.0/(np.linalg.norm(X0[:,i],normalize))
            X[:,i] = Mreg[i]*X0[:,i]
    else: X = X0

    # Lipschitz constant of gradient of smooth part of loss function
    L = np.linalg.norm(X.T.dot(X),2)+lam2

    # Now loop until converged or max iterations
    for iters in range(0, maxit):
         
        # Update w
        z = w + iters/float(iters+1)*(w - w_old)
        z = z - X.T.dot(X.dot(z)-Y)/L - lam2*z/L
        w = np.multiply(np.sign(z), (abs(z)-lam1/L).clip(0))

        # break condition based on convergence here.
        if np.linalg.norm(w-w_old, np.Inf) < tol: break
        w_old = w
    
    # Now that we have the sparsity pattern, used least squares.
    for j in range(m):
        biginds = np.where(w[:,j] != 0)[0]
        if len(biginds) != 0: w[biginds,j] = np.linalg.lstsq(X[:, biginds],Y[:,j])[0]

    # Finally, reverse the regularization so as to be able to use with raw data
    if normalize != 0:
        w_renorm = np.multiply(Mreg,w)
        return np.multiply(w_renorm, np.sign((abs(w_renorm)-hard_thresh).clip(0)))
    else: 
        return np.multiply(w, np.sign((abs(w)-hard_thresh).clip(0)))
```

```python
N = 720

P12 = []
P11 = []
P22 = []
P21 = []
P23 = []
P33 = []
P31 = []
P32 = []
count = 0

threshold = 5

for key in Training:
    
    print(key)
    
    runs = sarc_list[key]
    
    n = len(runs[0])
    
    state_fracs_ave = N*np.vstack([np.mean(np.vstack(np.array(run['xb_fraction_free']) for run in runs),axis=0),
                         np.mean(np.vstack(np.array(run['xb_fraction_loose']) for run in runs),axis=0),
                         np.mean(np.vstack(np.array(run['xb_fraction_tight']) for run in runs),axis=0)])
    
    state_fracs_min = N*np.vstack([np.min(np.vstack(np.array(run['xb_fraction_free']) for run in runs),axis=0),
                         np.min(np.vstack(np.array(run['xb_fraction_loose']) for run in runs),axis=0),
                         np.min(np.vstack(np.array(run['xb_fraction_tight']) for run in runs),axis=0)])
    
    P12_ave = np.mean(np.vstack(run['xb_trans_12']/(N*run['xb_fraction_free']) for run in runs), axis = 0)
    P21_ave = np.mean(np.vstack(run['xb_trans_21']/(N*run['xb_fraction_loose']) for run in runs), axis = 0)
    P23_ave = np.mean(np.vstack(run['xb_trans_23']/(N*run['xb_fraction_loose']) for run in runs), axis = 0)
    P31_ave = np.mean(np.vstack(run['xb_trans_31']/(N*run['xb_fraction_tight']) for run in runs), axis = 0)
    P32_ave = np.mean(np.vstack(run['xb_trans_32']/(N*run['xb_fraction_tight']) for run in runs), axis = 0)
        
    # Loop over timesteps
    for j in range(1,n):
        
        perm = runs[0]['actin_permissiveness'][j]
        zt = runs[0]['z_line'][j] - runs[0]['z_line'][j-1]
        z = runs[0]['z_line'][j]

        # Only look at ones where we can get transition probabilities
        if state_fracs_ave[0,j] > threshold and state_fracs_min[0,j] > 0:
            P12.append([perm, zt, z, P12_ave[j]])

        if state_fracs_ave[1,j] > threshold and state_fracs_min[1,j] > 0:
            P21.append([perm, zt, z, P21_ave[j]])
            P23.append([perm, zt, z, P23_ave[j]])

        if state_fracs_ave[2,j] > threshold and state_fracs_min[2,j] > 0:
            P31.append([perm, zt, z, P31_ave[j]])
            P32.append([perm, zt, z, P32_ave[j]])
```
