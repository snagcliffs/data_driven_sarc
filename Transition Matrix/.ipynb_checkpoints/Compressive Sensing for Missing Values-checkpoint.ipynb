```python
import sys

# Correct for use on various computers
sys.path.append('/home/samuel/Dropbox (uwamath)/muscle/MuscleModel/multifil/')
sys.path.append('/home/samuel/Dropbox (uwamath)/muscle/MuscleModel/Python_Codes')
sys.path.append('/Users/samuelrudy/Dropbox (uwamath)/muscle/MuscleModel/Python_Codes')

from multifil.aws.run import *
%pylab inline
pylab.rcParams['figure.figsize'] = (15,15)
import numpy as np
from scipy import sparse
import itertools
import operator
import pandas as pd
from zipfile import ZipFile
import ujson
from scipy.integrate import odeint
import pywt
import seaborn

def MovingAve(x, m):
    n = len(x)
    xs = np.zeros(n-2*m)
    for j in range(n-2*m):
        xs[j] = np.mean(x[j:j+2*m])
    return xs

def ProjZeroOne(x):
    condlist = [x<0, (x>0) & (x<=1), x>1]
    choicelist = [0,x, 1]
    return np.select(condlist, choicelist)

def Lasso(X0, Y, lam, ignore = [], w = np.array([0]), maxit = 100, normalize = 2, tol = 10**-3):
    """
    Uses accelerated proximal gradient (FISTA) to solve Lasso
    argmin (1/2)*||Xw-Y||_2^2 + lam||w||_1
    """
    
    # Obtain size of X
    n,d = X0.shape
    X = np.zeros((n,d), dtype=np.complex64)
    Y = Y.reshape(n,1)
    
    # Create w if none is given
    if w.size != d:
        w = np.zeros((d,1), dtype=np.complex64)
    w_old = np.zeros((d,1), dtype=np.complex64)
        
    # Initialize a few other parameters
    converge = 0
    objective = np.zeros((maxit,1))
    
    # First normalize data
    if normalize != 0:
        Mreg = np.zeros((d,1))
        for i in range(0,d):
            Mreg[i] = 1.0/(np.linalg.norm(X0[:,i],normalize))
            X[:,i] = Mreg[i]*X0[:,i]
    else: X = X0

    # Lipschitz constant of gradient of smooth part of loss function
    L = np.linalg.norm(X.T.dot(X),2)
    
    # Now loop until converged or max iterations
    for iters in range(0, maxit):
         
        # Update w
        z = w + iters/float(iters+1)*(w - w_old)
        z = z - X.T.dot(X.dot(z)-Y)/L
        for j in range(d): 
            if j not in ignore: w[j] = np.multiply(np.sign(z[j]), np.max([abs(z[j])-lam/L,0]))
            else: w[j] = z[j]

        # break condition based on convergence here.
        if np.linalg.norm(w-w_old, np.Inf) < tol: break
        w_old = w
    
    # Now that we have the sparsity pattern, used least squares.
    biginds = np.where(w != 0)[0]
    if biginds != []: w[biginds] = np.linalg.lstsq(X[:, biginds],Y)[0]

    # Finally, reverse the regularization so as to be able to use with raw data
    if normalize != 0: return np.multiply(Mreg,w)
    else: return w
    
def SoftThresh(x, lam):
    return np.multiply(np.sign(x), (abs(x)-lam).clip(0))
    
def LassoHuber(X0, Y, lam, gamma = 0.01, ignore = [], w = np.array([0]), maxit = 100, normalize = 2, tol = 10**-3):
    """
    Uses accelerated proximal gradient (FISTA) to solve Lasso with Huber penalty for error
    argmin_w ||Xw-Y||_H + lam||Pw||_1
    """
    
    # Obtain size of X
    n,d = X0.shape
    X = np.zeros((n,d), dtype=np.complex64)
    Y = Y.reshape(n,1)
    
    # Create w if none is given
    if w.size != d:
        w = np.zeros((d,1), dtype=np.complex64)
    w_old = np.zeros((d,1), dtype=np.complex64)
        
    # Initialize a few other parameters
    converge = 0
    objective = np.zeros((maxit,1))
    
    # First normalize data
    if normalize != 0:
        Mreg = np.zeros((d,1))
        for i in range(0,d):
            Mreg[i] = 1.0/(np.linalg.norm(X0[:,i],normalize))
            X[:,i] = Mreg[i]*X0[:,i]
    else: X = X0

    # Lipschitz constant of gradient of smooth part of loss function
    L = np.linalg.norm(X.T.dot(X),2)
    
    # Now loop until converged or max iterations
    for iters in range(0, maxit):
         
        # Update w
        z = w + iters/float(iters+1)*(w - w_old)
        G = X.dot(z)-Y
        z = z - X.T.dot(G - SoftThresh(G, gamma))/L
        for j in range(d): 
            if j not in ignore: w[j] = np.multiply(np.sign(z[j]), np.max([abs(z[j])-lam/L,0]))
            else: w[j] = z[j]

        # break condition based on convergence here.
        if np.linalg.norm(w-w_old, np.Inf) < tol: break
        w_old = w
    
    # Now that we have the sparsity pattern, used least squares.
    biginds = np.where(w != 0)[0]
    if biginds != []: w[biginds] = np.linalg.lstsq(X[:, biginds],Y)[0]

    # Finally, reverse the regularization so as to be able to use with raw data
    if normalize != 0: return np.multiply(Mreg,w)
    else: return w

def L1_MissingValues(f, period = None, basis = None, maxit = 10, l0_penalty = 0, regression = '2'):
    """
    Input:
        f = f(x) = series with missing values (should be stored as NaN)
        basis = matrix in C^(n x p) forming the basis we assume f(x) lies in 
        ignore = any frequencies we do not want to penalize
    """
    
    n = len(f)
    datapoints = np.where(~np.isnan(f))[0]
    if period == None: period = n
    
    # Create a basis matrix
    if basis is None: 
        t = np.arange(n)*2*np.pi/period
        basis = np.hstack(np.exp(1j*j*t).reshape((n,1)) for j in range(period))
        
    # Split into training and testing
    training = np.random.choice(datapoints, int(0.8*len(datapoints)), replace = False)
    testing = [j for j in datapoints if j not in training]
    
    # Training data
    training_basis = np.vstack([basis[j,:] for j in training])
    training_f = f[training].reshape((len(training),1))
    
    # Testing data
    testing_basis = np.vstack([basis[j,:] for j in testing])
    testing_f = f[testing].reshape((len(testing),1))
    
    # Set up stuff for searching for best lambda
    err_best = np.Inf
    Lam = 0.5*np.linspace(0, np.linalg.norm(training_basis.T.dot(training_f - mean(training_f)), np.Inf), maxit)
    
    # Solve to get full dataset
    for j in range(maxit):
        
        # Get a new predictor
        lam = Lam[j]
        
        if regression == '2': w = Lasso(training_basis, training_f, lam, ignore = [0,1])
        else: w = LassoHuber(training_basis, training_f, lam, ignore = [0,1])
            
        l0_err = 0#n*0.001*l0_penalty*np.count_nonzero(w)
        mean_err = 0#n*0.001*abs(np.mean(testing_f)-np.mean(real(testing_basis.dot(w))))
        
        err = np.linalg.norm(real(testing_basis.dot(w) - testing_f), 2) + l0_err + mean_err
        
        if err > err_best: return real(basis.dot(w_old))
        err_best = err        
        w_old = w
    
    return real(basis.dot(w))

def TwoOfThree(X_true, X_pred):
    
    T1,T2,T3 = X_true
    P1,P2,P3 = X_pred
    
    datapoints1 = np.where(~np.isnan(T1))[0]
    datapoints2 = np.where(~np.isnan(T2))[0]
    datapoints3 = np.where(~np.isnan(T3))[0]
    
    e1 = np.linalg.norm(T1[datapoints1] - P1[datapoints1].reshape(len(datapoints1)))
    e2 = np.linalg.norm(T2[datapoints2] - P2[datapoints2].reshape(len(datapoints2)))
    e3 = np.linalg.norm(T3[datapoints3] - P3[datapoints3].reshape(len(datapoints3)))
    
    if e1 > e2 and e1 > e3:
        return [1-P2-P3, P2, P3]
    elif e2 > e1 and e2 > e3:
        return [P1, 1-P1-P3, P3]
    else:
        return [P1, P2, 1-P1-P2]
    
def OneOfTwo(X_true, X_pred):
    
    T1,T2 = X_true
    P1,P2 = X_pred
    
    datapoints1 = np.where(~np.isnan(T1))[0]
    datapoints2 = np.where(~np.isnan(T2))[0]
    
    e1 = np.linalg.norm(T1[datapoints1] - P1[datapoints1].reshape(len(datapoints1)))
    e2 = np.linalg.norm(T2[datapoints2] - P2[datapoints2].reshape(len(datapoints2)))
    
    if e1 > e2:
        return [1-P2, P2]
    else:
        return [P1, 1-P1]
```
